```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

img1=mpimg.imread('/Users/joshking/Documents/01AI/Nike1.jpeg')
img2=mpimg.imread('/Users/joshking/Documents/01AI/Nike2.jpeg')
img3=mpimg.imread('/Users/joshking/Documents/01AI/Ellesse1.jpeg')
img4=mpimg.imread('/Users/joshking/Documents/01AI/Ellesse2.jpeg')
img5=mpimg.imread('/Users/joshking/Documents/01AI/Ellesse3.jpeg')
img6=mpimg.imread('/Users/joshking/Documents/01AI/Adidas1.jpeg')
img7=mpimg.imread('/Users/joshking/Documents/01AI/Adidas2.jpeg')
img8=mpimg.imread('/Users/joshking/Documents/01AI/Adidas3.jpeg')

catalogue = np.array([[img1, 'nike black sports sweater'], [img2, 'nike blue sports sweater'],
                 [img3, 'ellesse black sports sweater'], [img4, 'ellesse striped sports sweater'],
                 [img5, 'ellesse grey sports sweater'], [img6, 'adidas black sports sweater'],
                 [img7, 'adidas brown sports sweater'], [img8, 'adidas black sports sweater']])

def key_word(word, array):
    word = word
    array=array
    index_array=np.zeros(8)
    fig=plt.figure(figsize=(40, 40))
    for i in range(len(array)):
        if word in array[i,1]:
            index_array[i]=1
    k=0
    for j in range(len(index_array)):
        if index_array[j]==1:
            fig.add_subplot(len(array), len(array), k+1)
            plt.imshow(array[j,0])
            k+=1
            
    return fig

#imgplot = plt.imshow(img8)
#need to add in capabilities for capital letters, grammar, etc... Don't need to, 
#but mention this as easily being done
#Maybe get some images which better demonstrate the input being 1st hand, and output being second hand. 
```
